```python
# Importar bibliotecas necessárias
import pandas as pd
import matplotlib.pyplot as plt

# Importar bibliotecas necessárias
import pandas as pd
import matplotlib.pyplot as plt

# Carregar os dados do arquivo CSV
df = pd.read_excel("dados_desenvolvimento.xlsx")

# Realizar análise exploratória
total_horas_trabalhadas = df['Horas Trabalhadas'].sum()
media_horas_trabalhadas = df['Horas Trabalhadas'].mean()
total_bugs_corrigidos = df['Bugs Corrigidos'].sum()
media_bugs_corrigidos = df['Bugs Corrigidos'].mean()
total_tarefas_concluidas = df['Tarefas Concluídas'].sum()
media_tarefas_concluidas = df['Tarefas Concluídas'].mean()
df['Produtividade'] = df['Tarefas Concluídas'] / df['Horas Trabalhadas']
media_produtividade = df['Produtividade'].mean()

# Apresentar os resultados
print("Resultados da Análise Exploratória:")
print(f"Total de Horas Trabalhadas: {total_horas_trabalhadas} horas")
print(f"Média Diária de Horas Trabalhadas: {media_horas_trabalhadas:.2f} horas/dia")
print(f"Total de Bugs Corrigidos: {total_bugs_corrigidos}")
print(f"Média Diária de Bugs Corrigidos: {media_bugs_corrigidos:.2f} bugs/dia")
print(f"Total de Tarefas Concluídas: {total_tarefas_concluidas}")
print(f"Média Diária de Tarefas Concluídas: {media_tarefas_concluidas:.2f} tarefas/dia")
print(f"Produtividade Diária (Tarefas Concluídas por Hora): {media_produtividade:.2f} tarefas/hora")

# Visualizar os resultados
# 1. Gráfico de barras para Horas Trabalhadas por Dia
plt.figure(figsize=(10, 6))
plt.bar(df['Dia'], df['Horas Trabalhadas'], color='skyblue')
plt.xlabel('Dia')
plt.ylabel('Horas Trabalhadas')
plt.title('Horas Trabalhadas por Dia')
plt.xticks(rotation=45)
plt.show()

# 2. Gráfico de barras para Bugs Corrigidos por Dia
plt.figure(figsize=(10, 6))
plt.bar(df['Dia'], df['Bugs Corrigidos'], color='lightgreen')
plt.xlabel('Dia')
plt.ylabel('Bugs Corrigidos')
plt.title('Bugs Corrigidos por Dia')
plt.xticks(rotation=45)
plt.show()

# 3. Gráfico de barras para Tarefas Concluídas por Dia
plt.figure(figsize=(10, 6))
plt.bar(df['Dia'], df['Tarefas Concluídas'], color='orange')
plt.xlabel('Dia')
plt.ylabel('Tarefas Concluídas')
plt.title('Tarefas Concluídas por Dia')
plt.xticks(rotation=45)
plt.show()

# 4. Gráfico de linha para Produtividade Diária
plt.figure(figsize=(10, 6))
plt.plot(df['Dia'], df['Produtividade'], marker='o', color='purple')
plt.xlabel('Dia')
plt.ylabel('Produtividade (Tarefas/Hora)')
plt.title('Produtividade Diária')
plt.xticks(rotation=45)
plt.grid(True)
plt.show()
```
